```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

'''
Your Code
'''
import requests
import json
r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text
#len(response)
data = json.loads(response)
data
#print('å…±',len(data),'ç­†è³‡æ–™')
#for key in data[0]:
#    print(key)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

'''
Your Code
'''
for d in data:
    print('æ¨™é¡Œï¼š',d['title'])
    print('è²¼æ–‡æ™‚é–“ï¼š',d['createdAt'])
    print('ç•™è¨€äººæ•¸ï¼š',d['commentCount'])
    print('æŒ‰è®šäººæ•¸ï¼š',d['likeCount'])
    print('=====================================')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

'''
Your Code
'''
popular = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
popular_data = json.loads(popular.text)
totalTopicCount = int(len(popular_data))
totalCommentCount = 0
totalLikeCount = 0
for d in popular_data:
    totalCommentCount += int(d['commentCount'])
    totalLikeCount += int(d['likeCount'])
    
print('æ–‡ç« ç¸½æ•¸ï¼š',totalTopicCount)
print('ç•™è¨€ç¸½æ•¸ï¼š',totalCommentCount)
print('æŒ‰è®šç¸½æ•¸ï¼š',totalLikeCount)
print('å¹³å‡ç•™è¨€äººæ•¸ï¼š',int(totalCommentCount/totalTopicCount))
print('å¹³å‡æŒ‰è®šäººæ•¸ï¼š',int(totalLikeCount/totalTopicCount))
```
