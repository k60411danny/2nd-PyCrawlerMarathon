## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
import json
r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text
#len(response)
data = json.loads(response)
data
#print('共',len(data),'筆資料')
#for key in data[0]:
#    print(key)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
for d in data:
    print('標題：',d['title'])
    print('貼文時間：',d['createdAt'])
    print('留言人數：',d['commentCount'])
    print('按讚人數：',d['likeCount'])
    print('=====================================')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''
popular = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
popular_data = json.loads(popular.text)
totalTopicCount = int(len(popular_data))
totalCommentCount = 0
totalLikeCount = 0
for d in popular_data:
    totalCommentCount += int(d['commentCount'])
    totalLikeCount += int(d['likeCount'])
    
print('文章總數：',totalTopicCount)
print('留言總數：',totalCommentCount)
print('按讚總數：',totalLikeCount)
print('平均留言人數：',int(totalCommentCount/totalTopicCount))
print('平均按讚人數：',int(totalLikeCount/totalTopicCount))
```
